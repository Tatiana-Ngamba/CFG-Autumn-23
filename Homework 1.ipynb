### Overview
### - In order for me to find all the missing values in the Titanic data, I have used the **'isnull()'** function. I have also combined this with the **'sum'** function to count all the missing values per column.
### - When I run titanic.isnull(), it returns a DataFrame with **'True'**  and **'False'**  values. I can see 'True' in the **'age' & 'deck'**  column - which indicates that, these this is where a value is missing.
### -  **titanic.isnull()** returns a DataFrame in the same number of rows and columns as the **titanic dataset** with **True** where a value is missing and **False** otherwise.
### -  **.sum()** is then used to sum up these Boolean values per column, where **True** is treated as **1** and **False** as **0**. This gives the total **count of missing values** in each column.
### -  As a result of the missing values is a Pandas Series where the index represents the column names, and the values represent the total count of the missing values per column.
