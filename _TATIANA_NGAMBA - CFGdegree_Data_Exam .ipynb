1. PANDAS (25 points)

```python
#Importing libraries
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# 1 (2 marks). Load the netflix_titles.csv dataset as a Pandas dataframe:
df = pd.read_csv('data/netflix_titles.csv')
df
```

```python
# 2 (2 marks). Show the first 10 rows of the dataset
df.head(10)
```

```python
# 3 (1 mark). Show the last 5 rows
df.tail(5)
```

```python
# 4 (4 marks). Count the number of rows which have no missing value in any of the columns, and show the resulting data frame.

number_of_rows = df.dropna().shape[0]

print("The number of rows without missing values:", number_of_rows)

# The resulting DataFrame
df_not_missing = df.dropna()
df_not_missing
```

```python
# 5 (3 marks). Drop everything that is not a 'Movie' and show the resulting data frame.

movie_df = df[df['type'] == 'Movie']
movie_df
```

```python
# checking Data types
movie_df.dtypes
```

```python
# 6 (7 marks). For every movie where the duration is missing, replace it with the average
  # duration of the movies in the dataset.

# Converting 'duration'
movie_df['duration'] = movie_df['duration'].astype(str).str.extract('(\d+)', expand=False)

# Converting the extracted values to float
movie_df['duration'] = pd.to_numeric(movie_df['duration'], errors='coerce')

# Calculating the mean duration
average_duration = movie_df['duration'].mean()


# Replacing missing durations values
movie_df['duration'].fillna(average_duration, inplace=True)
df
```

```python
# checking Data types again
movie_df.dtypes
```

```python
# 7 (3 marks). Are there any duplicates in the data? If yes, please remove them.
movie_df.drop_duplicates()
```

```python
# 8 (2 marks). Rename all columns to start with a capital letter.
movie_df.rename(index=None, columns=str.title,inplace=True)
movie_df
```

```python
# 9 (1 mark). Save the cleaned dataset as a new csv file.
movie_df.to_csv('data/cleaned_movie_data.csv')
```

2. MATPLOTLIB (25 points)

Use the netflix dataset to solve the exercises in this section.

```python
#Importing libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movie_df.head(10)
```

```python
# 1 (10 marks). Create a visualisation which shows the frequency of movies by year of release.
# counting the value
movie_by_year = movie_df['Release_Year'].value_counts().sort_index()
movie_by_year
```

```python
# Creating a bar plot

plt.figure(figsize=(12, 8))
plt.bar(movie_by_year.index, movie_by_year.values, color='blue')
plt.title('Frequency of Movies by Year of Release', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(axis='y', alpha=1, linestyle='--', linewidth=1)

plt.show()
```

```python
# 2 (15 marks). Create a pie chart which shows the 5 top most popular genres of the movis
  # in the dataset.
#Using Groupby and Calculating movie genre
top_5_movies_counts = movie_df.groupby('Listed_In')['Title'].count().sort_values(ascending=False).head(5)
top_5_movies_counts
```

```python
# Defining colours for each movie genre
cols = ['royalblue','purple','gold','green','orchid']

# Creating a pie chat
plt.figure(figsize=(10, 8))
plt.pie(x = top_5_movies_counts, 
        labels=top_5_movies_counts.index, 
        colors=cols,
        shadow=True,
        startangle=90,
        explode = (0, 0.1, 0, 0, 0),
        autopct='%1.1f%%')
plt.title('Top 5 Most Popular Genres', fontweight='bold')
plt.legend(top_5_movies_counts.index, title='Genres', loc='best', prop={'size': 8})
plt.show()
```
